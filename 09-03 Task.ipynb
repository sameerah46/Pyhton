```python
print(df.duplicated().sum())
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df.dtypes)
```

```python
df['price'] = df['price'].astype(float)
print(df['price'].dtype)
```

```python
df['region'] = df['region'].str.strip().str.title()
print(df['region'])
```

```python
# Total sales by region
total_sales = df.groupby("region")["sales"].sum()
print("Total Sales by Region:\n", total_sales)

# Region with highest sales
highest_region = total_sales.idxmax()
print("\nRegion with highest sales:", highest_region)
```

```python
df.groupby("product")["discount"].mean()
```

```python
df.groupby("product")["quantity"].sum().sort_values(ascending=False).head(5)
```

```python
df['month'] = df['date'].dt.month

# Calculate total sales per month
monthly_sales = df.groupby("month")["sales"].sum()
print("Monthly Sales Trend:\n", monthly_sales)
```

```python
avg_with_discount = df[df['discount'] > 0]["sales"].mean()

avg_without_discount = df[df['discount'] == 0]["sales"].mean()

print("Average Sales with Discount:", avg_with_discount)
print("Average Sales without Discount:", avg_without_discount)
```

```python
df.groupby("product")["sales"].sum().idxmax()
```

```python
# Total sales by customer type
total_sales_customer = df.groupby("customer_type")["sales"].sum()
print("Total Sales by Customer Type:\n", total_sales_customer)

# Average sales per order by customer type
avg_sales_customer = df.groupby("customer_type")["sales"].mean()
print("\nAverage Sales per Order by Customer Type:\n", avg_sales_customer)

# Count of orders by customer type
order_count = df.groupby("customer_type")["OrderID"].count()
print("\nNumber of Orders by Customer Type:\n", order_count)
```
